```python
#Link dos arquivo csv transformado em variável para facilitar a leitura
csv2020Expo = "https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncm/EXP_2020.csv"

#Importação do arquivo CSV utilizando ; como separador, sem isso a tabela não é importada corretamente.
EXP_2020=pd.read_csv(csv2020Expo,sep=';')
```

A tabela parece muito bem organizada, não aparenta ter valores ausentes/nulos e contêm a maioria de seus atributos com o tipo int. Agora farei o mesmo processo para as demais por meio de função.

```python
#Exportando demais tabelas
csv2021Expo = "https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncm/EXP_2021.csv"

csv2020Imp = "https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncm/IMP_2020.csv"
csv2021Imp = "https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncm/IMP_2021.csv"

#Importação dos arquivo CSV utilizando ; como separador, sem isso a tabela não é importada corretamente.
EXP_2021=pd.read_csv(csv2021Expo,sep=';')
IMP_2020=pd.read_csv(csv2020Imp,sep=';')
IMP_2021=pd.read_csv(csv2021Imp,sep=';')
```

```python
def relatorio_tabelas(df,name=''):
  print(f'Relatório da tabela {name}')
  print("-" * 40)
  print('Shape:',df.shape)
  if 'EXP' in name:
    df.columns = ['Ano','Mês','Produto','Unidade','País','Estado','Transporte','Alfândega','Quantidade','Peso(Kg)','Valor']
  else:
    df.columns=['Ano','Mês','Produto','Unidade','País','Estado','Transporte','Alfândega','Quantidade','Peso(Kg)','Valor','Frete','Seguro']
  print(df.info())
  print("Valores nulos:\n", df.isnull().sum())
  print("Duplicadas:", df.duplicated().sum())
  print("Anos únicos:", df['Ano'].unique())
  print("UFs únicas:", df['Estado'].unique())
  print()

relatorio_tabelas(EXP_2021,'EXP_2021')
relatorio_tabelas(IMP_2020,'IMP_2020')
relatorio_tabelas(IMP_2021,'IMP_2021')
```

Como fica evidente nas análises realizadas, todas as tabelas estão sem valores faltantes ou duplicatas e estão, ao todo, bem estruturadas. Logo, podemos prosseguir com a análise.
